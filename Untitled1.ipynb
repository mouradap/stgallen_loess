```python
preprocessed_df = df.copy()
preprocessed_df['date'] = pd.to_datetime(preprocessed_df['date'])
# preprocessed_df['date'] = pd.to_datetime(preprocessed_df['date']) - pd.to_timedelta(7, unit = 'd')
```

```python
preprocessed_df = preprocessed_df[preprocessed_df['date'] >= '2020-08-02']
```

```python
preprocessed_df.fillna(0, inplace=True)
```

```python
symp_idx = ['2', '3', '4', '12', '14', '15']
```

```python
for symp in symp_idx:
    preprocessed_df[symp] = preprocessed_df[symp].astype(int)
```

```python
preprocessed_df['12'].unique()
```

```python
grouped_df = preprocessed_df.groupby(['userId', pd.Grouper(key='date', freq='W-SUN')])[symp_idx].first().reset_index().sort_values('date')
grouped_df
```

```python
symp_dict = {
    '2': 'sore_throat',
    '3': 'cough',
    '4': 'shortness_breath',
    '12': 'fever',
    '14': 'limb_muscle_pain',
    '15': 'loss_of_taste_smell'
}
symptom_list = [symp for symp in symp_dict.values()]
symptom_list.append('foph')
```

```python
def __replace_symptom(cols):
    for n, i in enumerate(cols):
        if i in symp_dict.keys():
            cols[n] = symp_dict[i]
    return cols
```

```python
cols = grouped_df.columns.to_list()
cols = __replace_symptom(cols)
grouped_df.columns = cols

grouped_df
```

```python
def __calc_foph(row):
    if row.fever == 1 or row.cough == 1 or row.shortness_breath == 1 or row.sore_throat == 1 or row.loss_of_taste_smell == 1:
        foph = 1
    else:
        foph = 0

    return foph
```

```python
grouped_df['foph'] = grouped_df.apply(__calc_foph, axis = 1)
```

```python
grouped_df
```

```python
def get_pos_ratio(symp):

    ratio_series = []
    for date in grouped_df['date'].unique():
        date_df = grouped_df[grouped_df['date'] == date]

        try:
            negatives, positives = date_df[symp].value_counts()
            ratio = positives / (positives + negatives)
            ratio = round(ratio, 4)
        except ValueError:
            ratio = 1.0

        ratio_series.append(ratio)
        
    return ratio_series
```

```python
grouped_df['date'].unique()
```

```python
ratio_df = pd.DataFrame()
ratio_df['date'] = grouped_df['date'].unique()

for symp in symptom_list:
    ratio_df[symp] = get_pos_ratio(symp)
```
